```python
from PIL import Image
```

```python
image2=Image.open('myphoto.jpg')
type(image2)
```

```python
display(image2)
print(image2.size)
```

```python
width=256
ratio=float(width)/image2.size[0]
height=int(image2.size[1]*ratio)
nim=image2.resize((width,height), Image.BILINEAR)
print(nim.size)
display(nim)
```

```python
width=256
height=256
nim=image2.resize((width,height), Image.BILINEAR)
print(nim.size)
display(nim)
```

```python
nim.save('HW1.jpg')
print(nim.mode,nim.format)
display(nim)
```
